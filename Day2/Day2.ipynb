```python
reports = []

with open('input.txt') as f:
    for line in f:
        report = list(map(int, line.rstrip('\n').split(" ")))
        reports.append(report)

reports
```

```python
safe_report_count = 0
safe_reports = []

for report in reports:
    unsafe = False 

    if report[1] == report[0]:
        continue
    if abs(report[1] - report[0]) > 3:
        continue

    increasing = True

    if report[1] < report[0]:
        increasing = False

    last_num = report[1]

    for num in report[2:]:
        if (abs(num - last_num)) > 3:
            unsafe = True
            
            break
        if increasing:
            if num <= last_num:
                unsafe = True
                break
        else:
            if num >= last_num:
                unsafe = True
                break
        
        last_num = num
        
    if not unsafe:
        print(f"{report} is safe")
        safe_report_count += 1
        safe_reports.append(report)
    


safe_report_count
        
    
```
